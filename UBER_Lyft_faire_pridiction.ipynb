**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

To get mean squared error and R-squared

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_predict)
print('mean squared error:',mse)

r2 = r2_score(y_test, y_predict)
print('R-squared:',r2)

print('train accuracy',model.score(X_train,y_train)*100)
print('test accuracy',model.score(X_test,y_test)*100)
```

**Ridge**

```python
# Initialize Ridge model with a specific alpha (regularization strength)
ridge = Ridge(alpha=1.0)

# Train the model
ridge.fit(X_train2, y_train2)

# Make predictions
y_pred_ridge = ridge.predict(X_test2)

# Evaluate the model
mse_ridge = mean_squared_error(y_test2, y_pred_ridge)
print(f"Ridge Mean Squared Error:",mse_ridge)

ridge_R2_score = r2_score(y_test2,y_pred_ridge)

print("R-squared score:", ridge_R2_score)
print('train accuracy',ridge.score(X_train2,y_train2)*100)
print('test accuracy',ridge.score(X_test2,y_test2)*100)
```
